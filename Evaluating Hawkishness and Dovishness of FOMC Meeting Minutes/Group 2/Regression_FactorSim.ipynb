```python
def merge_market_data(simimarity_factor, market_data):
    # Iterate over each DataFrame in market_data
    for market_df in market_data:
        # Select columns that end with '_change' or '_%change'
        selected_columns = [col for col in market_df.columns if col.endswith('_change') or col.endswith('_%change')]
        
        # Add the 'Date' column to the selected columns to use it for merging
        selected_columns = ['Date'] + selected_columns
        
        # Perform an inner merge on 'date' from simimarity_factor and 'Date' from market_df
        simimarity_factor = pd.merge(simimarity_factor, market_df[selected_columns], left_on='date', right_on='Date', how='inner')
        
        # Drop the redundant 'Date' column after merging
        simimarity_factor = simimarity_factor.drop(columns=['Date'])
    
    return simimarity_factor
```

```python
market_data = [gt10_df, gt2_df, gtspread_df, gold_df, VIX_df, SP500_df]
minutes_regression_data = merge_market_data(minutes_sim,market_data)
presconf_regression_data = merge_market_data(presconf_sim,market_data)
speech_regression_data = merge_market_data(speech_sim,market_data)
```

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

def format_significance(p_value):
    """Format significance level based on p-value."""
    if p_value < 0.01:
        return "***"
    elif p_value < 0.05:
        return "**"
    elif p_value < 0.1:
        return "*"
    else:
        return ""

def perform_regression_for_dataset(regression_data, label):
    # Initialize a dictionary to collect results
    results_dict = {}
    
    # Identify all columns ending with '_change' or '_%change'
    target_columns = [col for col in regression_data.columns if col.endswith('_change') or col.endswith('_%change')]

    # Group columns by their base name before '_change' and '_%change'
    base_names = sorted(set(col.rsplit('_', 1)[0] for col in target_columns))
    
    # Initialize the result dictionary for this dataset
    results_dict[label] = {}
    
    # Loop through each base name (e.g., GT10, GT2, etc.)
    for base_name in base_names:
        # Extract the columns for '_change' and '_%change'
        change_col = f"{base_name}_change"
        pct_change_col = f"{base_name}_%change"
        
        # Only process if both columns exist
        if change_col in regression_data.columns and pct_change_col in regression_data.columns:
            # Create sub-dict for this base name
            results_dict[label][base_name] = {}
            
            # Define X (independent variables) as 'int score' and 'inf score'
            X = regression_data[['int score', 'inf score']]
            X = sm.add_constant(X)  # Add intercept
            
            # Loop through change and %change columns
            for col_name in [change_col, pct_change_col]:
                y = regression_data[col_name]
                
                # Perform the regression
                model = sm.OLS(y, X).fit()
                
                # Collect coefficients and standard errors
                coeff_int = model.params['int score']
                coeff_inf = model.params['inf score']
                stderr_int = model.bse['int score']
                stderr_inf = model.bse['int score']
                
                # Collect p-values to mark significance
                p_value_int = model.pvalues['int score']
                p_value_inf = model.pvalues['inf score']
                
                # Format coefficients with significance
                int_score_result = f"{coeff_int:.4f}{format_significance(p_value_int)}"
                inf_score_result = f"{coeff_inf:.4f}{format_significance(p_value_inf)}"
                
                # Store results for both int score and inf score
                results_dict[label][base_name][col_name] = {
                    'int_score': (int_score_result, f"({stderr_int:.4f})"),
                    'inf_score': (inf_score_result, f"({stderr_inf:.4f})"),
                    'n_obs': len(y),
                    'r_squared': model.rsquared
                }

    # After collecting all the results, prepare the final DataFrame
    rows = []
    index_labels = []
    
    # Loop over the rows (int score and inf score for this dataset)
    for score_type in ['int_score', 'inf_score']:
        row_coefficients = []
        row_std_errors = []
        index_labels.append(f"{label} {score_type} Coefficients")
        index_labels.append(f"{label} {score_type} Std Errors")
        
        for base_name in base_names:
            for change_type in ['_change', '_%change']:
                change_col = f"{base_name}{change_type}"
                
                if base_name in results_dict[label] and change_col in results_dict[label][base_name]:
                    result = results_dict[label][base_name][change_col]
                    # First row: coefficients with significance
                    row_coefficients.append(result[score_type][0])  # Coefficient with significance
                    # Second row: standard errors
                    row_std_errors.append(result[score_type][1])  # Standard Error
                else:
                    row_coefficients.append(np.nan)
                    row_std_errors.append(np.nan)
        
        # Append both rows (coefficients and standard errors)
        rows.append(row_coefficients)
        rows.append(row_std_errors)
    
    # Add R² and sample size (n_obs) rows
    r_squared_row = []
    sample_size_row = []
    index_labels.append(f"{label} R²")
    index_labels.append(f"{label} Sample Size")
    
    for base_name in base_names:
        for change_type in ['_change', '_%change']:
            change_col = f"{base_name}{change_type}"
            
            if base_name in results_dict[label] and change_col in results_dict[label][base_name]:
                result = results_dict[label][base_name][change_col]
                # R-squared and sample size
                r_squared_row.append(f"{result['r_squared']:.4f}")
                sample_size_row.append(f"{result['n_obs']}")
            else:
                r_squared_row.append(np.nan)
                sample_size_row.append(np.nan)
    
    # Append R² and sample size rows
    rows.append(r_squared_row)
    rows.append(sample_size_row)
    
    # Create the column names (multi-level)
    columns = pd.MultiIndex.from_product([base_names, ['_change', '_%change']], names=['Variable', 'Type'])
    
    # Convert results to DataFrame
    results_df = pd.DataFrame(rows, index=index_labels, columns=columns)
    
    # Return the DataFrame for this dataset
    return results_df

# Example usage for each dataset
minutes_regression_results = perform_regression_for_dataset(minutes_regression_data, 'Minutes')
presconf_regression_results = perform_regression_for_dataset(presconf_regression_data, 'Press Conference')
speech_regression_results = perform_regression_for_dataset(speech_regression_data, 'Speech')
```
