```python
import pandas as pd
import statsmodels.api as sm
```

```python
index_df = pd.read_csv('./data/FOMC_RoBERTa_sentiment_index.csv', parse_dates=['Date'])
index_df = index_df.sort_values('Date')
index_df = index_df[index_df['Number of Sentences']>3]
index_df = index_df.groupby(['File Type','Date'])[['Sentiment Index']].mean()
index_df.reset_index(inplace=True)
index_df
```

```python
minutes_idx = index_df[index_df['File Type']=='FOMC_Minutes']
minutes_idx
```

```python
presconf_idx = index_df[index_df['File Type']=='FOMC_Presconf']
presconf_idx
```

```python
speech_idx = index_df[index_df['File Type']=='FOMC_Speech']
speech_idx
```

# Regression

```python
def format_significance(p_value):
    """Format significance level based on p-value."""
    if p_value < 0.01:
        return "***"
    elif p_value < 0.05:
        return "**"
    elif p_value < 0.1:
        return "*"
    else:
        return ""
```

```python
# Define a dictionary mapping the name of the index datasets to their corresponding DataFrames
idx_dict = {
    'minutes_idx': minutes_idx,
    'presconf_idx': presconf_idx,
    'speech_idx': speech_idx
}
# Define a list of market datasets
market_lst = [gt10_df, gt2_df, gtspread_df, gold_df, VIX_df, SP500_df]

results_df = pd.DataFrame()
for idx_name, idx_data in idx_dict.items():
    for market_data in market_lst:
        # Merge the market and index data on the 'Date' column
        df = pd.merge(market_data, idx_data, on='Date', how='inner')
        # Filter the columns that end with '_change' or '_%change'
        change_columns = df.filter(regex='_change$|_%change$').columns
        for col in change_columns:
            X = df[['Sentiment Index']]
            X = sm.add_constant(X)
            y = df[col]
            model = sm.OLS(y, X).fit()

            # Extract the coefficient, standard error, and p-value for 'Sentiment Index'
            coeff_idx = model.params['Sentiment Index']
            stderr_idx = model.bse['Sentiment Index']
            p_value_idx  = format_significance(model.pvalues['Sentiment Index'])

            # Extract the coefficient, standard error, and p-value for the constant (intercept)
            coeff_const = model.params['const']
            stderr_const = model.bse['const']
            p_value_const = format_significance(model.pvalues['const'])

            # Calculate the R-squared value for the model
            r2 = model.rsquared

            new_row = pd.DataFrame({
                'Index Data': [idx_name],
                'Variable': col,
                'Constant Coefficient': f"{coeff_const:.4f}{p_value_const}",
                'Constant Std Error': f"({stderr_const:.4f})",
                'Sentiment Index Coefficient': f"{coeff_idx:.4f}{p_value_idx}",
                'Sentiment Index Std Error': f"({stderr_idx:.4f})",
                'R-squared': f"{r2:.4f}"
            })
            results_df = pd.concat([results_df, new_row], ignore_index=True)
results_df
results_df.to_csv('regression_result.csv', index=False)
```
